```python
import numpy as np
x1 = np.random.multivariate_normal(np.zeros(2), np.identity(2) + np.ones(2) @ np.ones(2).T, 1000)
```

```python
x2 = np.random.multivariate_normal(np.array((0, 0)), np.identity(2) + 0.2 * np.ones(2) @ np.ones(2).T, 1000)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x1[:,0], x1[:,1], c='red')
plt.scatter(x2[:,0], 10*x2[:,1], c='green')
```

```python
np.corrcoef(x1[:,0], x1[:,1])[0,1], np.corrcoef(x2[:,0], 10*x2[:,1])[0,1]
```

```python
x = np.concatenate([x1, x2], axis=0)
np.corrcoef(x[:,0], x[:,1])[0,1]
```
